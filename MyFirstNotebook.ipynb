```python
from pyspark.sql.functions import col

data = [
    (1, "Alice", 29),
    (2, "Bob", 31),
    (3, "Cathy", 25)
]

df = spark.createDataFrame(
    data,
    ["id", "name", "age"]
)

display(df)

df_with_age_plus_one = df.withColumn(
    "age_plus_one",
    col("age") + 1
)

display(df_with_age_plus_one)
```
